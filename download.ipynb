```python
import pandas as pd
import urllib.request
import os
from datetime import date
Downloading_List=pd.read_csv('par_to_download_20220817.txt',header=None)
```

```python
date_today= date.today().strftime("%Y%m%d")

DIR_OUT = "./Data/"+date_today+'/'
if not os.path.exists(DIR_OUT):
    os.makedirs(DIR_OUT)
print('Data Location: ',DIR_OUT)

for fi in range(0,len(Downloading_List[0])):
    url = Downloading_List[0][fi]
    filename =url.split('/')[-1]
    urllib.request.urlretrieve(url, DIR_OUT+filename)
    print(fi,'- Downloaded: ',filename)
```
